## P19– count( ) 

```python
#P19.1 Rosalind #6 Counting DNA Nucleotides
# Read DNA string (replace this with your sequence or file input)
s = input("Enter DNA string: ").strip()

# Count nucleotides
A = s.count('A')
C = s.count('C')
G = s.count('G')
T = s.count('T')

# Print results
print(A, C, G, T)
```

```python
# P19.2
# Input string from Rosalind problem # 5
s = "We tried list and we tried dicts also we tried Zen"

# Empty dictionary
counts = {}

# Split string into words
words = s.split()

for word in words:
    if word not in counts:       # only add new words
        counts[word] = words.count(word)

# Print results
for key, value in counts.items():
    print(key, value)
```

## P23– SeqUtils.gc_fraction( ) 

```python
# P23.1
# Import parse for reading fasta and gc_fraction for GC content
from Bio.SeqIO import parse
from Bio.SeqUtils import gc_fraction

# Loop through each record in the fasta file
for seq_record in parse("content/practice_9.txt", "fasta"):

    # Print the sequence ID
    print("ID:", seq_record.id)

    # Print the sequence
    print("Sequence:", seq_record.seq)

    # Print the length
    print("Length:", len(seq_record))

    # Print the GC content (as %)
    print("GC Content (%):", gc_fraction(seq_record.seq) * 100)

    # Print the full SeqRecord object
    print(seq_record)
    print("-" * 40)
```

```python
# P23.2  Rosalind #9  Computing GC Content

from Bio.SeqIO import parse
from Bio.SeqUtils import gc_fraction

# Open FASTA file
with open("content/rosalind_gc.txt", "r") as f:
    best_id, best_gc = None, -1.0
    
    for rec in parse(f, "fasta"):
        gc = gc_fraction(str(rec.seq)) * 100
        if gc > best_gc:
            best_id, best_gc = rec.id, gc

print(best_id)
print(f"{best_gc:.6f}")
```
